```python
import string
import random

def generate_password(length, include_uppercase=True, include_lowercase=True, include_numbers=True, include_symbols=True):
    character_set = ''

    if include_uppercase:
        character_set += string.ascii_uppercase

    if include_lowercase:
        character_set += string.ascii_lowercase

    if include_numbers:
        character_set += string.digits

    if include_symbols:
        character_set += string.punctuation

    password = ''.join(random.choice(character_set) for _ in range(length))
    return password

def evaluate_password_strength(password):
    strength = 0

    if len(password) >= 8:
        strength += 1

    if string.ascii_uppercase in password:
        strength += 1

    if string.ascii_lowercase in password:
        strength += 1

    if string.digits in password:
        strength += 1

    if string.punctuation in password:
        strength += 1

    return strength

def main():
    password_length = int(input("Enter desired password length: "))

    while True:
        include_uppercase = input("Include uppercase letters (y/n): ")
        if include_uppercase.lower() == 'y':
            include_uppercase = True
        elif include_uppercase.lower() == 'n':
            include_uppercase = False
        else:
            print("Invalid input. Please enter 'y' or 'n'.")
            continue

        include_lowercase = input("Include lowercase letters (y/n): ")
        if include_lowercase.lower() == 'y':
            include_lowercase = True
        elif include_lowercase.lower() == 'n':
            include_lowercase = False
        else:
            print("Invalid input. Please enter 'y' or 'n'.")
            continue

        include_numbers = input("Include numbers (y/n): ")
        if include_numbers.lower() == 'y':
            include_numbers = True
        elif include_numbers.lower() == 'n':
            include_numbers = False
        else:
            print("Invalid input. Please enter 'y' or 'n'.")
            continue

        include_symbols = input("Include symbols (y/n): ")
        if include_symbols.lower() == 'y':
            include_symbols = True
        elif include_symbols.lower() == 'n':
            include_symbols = False
        else:
            print("Invalid input. Please enter 'y' or 'n'.")
            continue

        break

    generated_password = generate_password(password_length, include_uppercase, include_lowercase, include_numbers, include_symbols)
    password_strength = evaluate_password_strength(generated_password)

    print("Generated password:", generated_password)
    print("Password strength:", password_strength)

if __name__ == "__main__":
    main()
```
